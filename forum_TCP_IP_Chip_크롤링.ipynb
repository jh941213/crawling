```python
import re

# 파일 경로
file_path = "deok.txt"

# 파일 열기
with open(file_path, "r", encoding="utf-8") as file:
    # 파일 내용 읽기
    file_content = file.read()

# 정규식을 이용하여 패턴에 맞는 모든 본문 가져오기
pattern = r'<a href="/t/topic/\d+".*?>(.*?)</a>'
matches = re.findall(pattern, file_content)
result = []
if matches:
    for match in matches:
        result.append(match)
else:
    print("해당 패턴에 맞는 본문을 찾을 수 없습니다.")
```
